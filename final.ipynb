```python
import import_ipynb
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt
from machineLearning import *
```

```python
dir_path = './the_end'
sources = []
for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.png' in file:
            file_path = os.path.join(root, file)
            sources.append(file_path)
            
src = [[] for _ in range(100)]
index = 0
temp = sources[0]
cnt = 0
while cnt < len(sources):
    if temp[10:-6] == sources[cnt][10:-6]:
        src[index].append(sources[cnt])
        temp = sources[cnt]
        cnt += 1
    else:
        temp = sources[cnt]
        index += 1
src = list(filter(None, src))

dir_path = './cropped'
org_sources = []
for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.jpg' in file:
            file_path = os.path.join(root, file)
            org_sources.append(file_path)
org = [[] for _ in range(100)]
index = 0
temp = org_sources[0]
cnt = 0

while cnt < len(org_sources):
    if temp[13:17] == org_sources[cnt][13:17]:
        org[index].append(org_sources[cnt])
        temp = org_sources[cnt]
        cnt += 1
    else:
        temp = org_sources[cnt]
        index += 1
org = list(filter(None, org))
for i in src:
    print(i)
```

```python
recg_img = [[] for _ in range(100)]
recg_num = []
index = 0
for data in src:
    for img in data:
        fname = './cropped/' + img[10:-6] + '.jpg'
        recg_img[index].append(fname)
        break
    index += 1
recg_img = list(filter(None, recg_img))

for data in src:
    res = ''
    for img in data:
        res += recognition(img)
    recg_num.append(res)


recg_num = list(filter(None, recg_num))

for y in range(len(recg_img)):
    for x in range(len(recg_img[y])):
        plt.figure(figsize = (4, 6))
        plt.title(recg_num[x+y])
        plt.imshow(cv2.imread(recg_img[y][x]))
        plt.axis('off')
```
