```python
import pandas as pd
import cv2
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import os
```

```python
dir_path = './train_data/new_data'
sources = []
for (root, directories, files) in os.walk(dir_path):
    for file in files:
        if '.png' in file:
            file_path = os.path.join(root, file)
            sources.append(file_path)

# print(sources)
src = [[] for _ in range(100)]
index = 0
temp = sources[0]
cnt = 0
while cnt < len(sources):
    if temp[-9:-7] == sources[cnt][-9:-7]:
        src[index].append(sources[cnt])
        temp = sources[cnt]
        cnt += 1
    else:
        temp = sources[cnt]
        index += 1
src = list(filter(None, src))
# for i in src:
#     print(i)
img_cnt = len(src[0])
```

```python
FNAME = 'digits.npz'

def machineLearning():
    img = cv2.imread('digits.png')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    cells = [np.hsplit(row,100) for row in np.vsplit(gray,50)]
    x = np.array(cells)
    train = x[:,:].reshape(-1,400).astype(np.float32)

    k = np.arange(10)
    train_labels = np.repeat(k,500)[:,np.newaxis]

    np.savez(FNAME,train=train,train_labels = train_labels)

def resize20(pimg):
    img = cv2.imread(pimg)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    grayResize = cv2.resize(gray,(20,20))
    ret, thresh = cv2.threshold(grayResize, 150, 255,cv2.THRESH_BINARY)
#     plt.imshow(thresh, 'gray')
    return thresh.reshape(-1,400).astype(np.float32)

def loadTrainData(fname):
    with np.load(fname) as data:
        train = data['train']
        train_labels = data['train_labels']

    return train, train_labels

def checkDigit(test, train, train_labels):
    knn = cv2.ml.KNearest_create()
    knn.train(train, cv2.ml.ROW_SAMPLE, train_labels)

    ret, result, neighbours, dist = knn.findNearest(test, k=5)

    return result

def recognition(img):
    loadTrainData(FNAME)
    train, train_labels = loadTrainData(FNAME)
    img = resize20(img)
    result = checkDigit(img, train, train_labels)
    rec_res = result[0][0]
    return str(int(rec_res))
    

train, train_labels = loadTrainData(FNAME)
print(train.shape)
print(train_labels.shape)

for i in range(10):
    for j in range(img_cnt):
        img = src[i][j]
#         plt.figure(figsize=(2,2))
#         plt.imshow(cv2.imread(img), 'gray')
        test = resize20(img)
        result = checkDigit(test, train, train_labels)
        if i != int(result[0][0]):
            print('Current Number : ', i)
            print('result = ',int(result[0][0]))
            k = int(input())
            train = np.append(train, test, axis = 0)
            newLabel = np.array(k).reshape(-1,1)
            train_labels = np.append(train_labels, newLabel, axis = 0)
            np.savez(FNAME, train = train, train_labels = train_labels)
            print('saved')
        else:
            print('Current Number : ', i)
            print('result = ',int(result[0][0]))
```
